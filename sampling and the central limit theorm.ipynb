```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#binomial(n,p,size)
pop1=np.random.binomial(10,0.2,10000)
pop2=np.random.binomial(10,0.5,10000)

plt.hist(pop1,alpha=0.3,label='pop1')
plt.hist(pop2,alpha=0.3,label='pop2')   
plt.legend(loc='upper right')
plt.show()
         
         
         
```

```python
sample1=np.random.choice(pop1,100,replace =True)
sample2=np.random.choice(pop2,100,replace=True)
plt.hist(sample1,alpha=0.3,label='sample1')
plt.hist(sample2,alpha=0.3,label='sample2')
plt.legend(loc='upper right')
plt.show()
```

```python
mean1=np.mean(sample1)
print(mean1)
std1=np.std(sample1)
print(std1)

mean2=np.mean(sample2)
print(mean2)
std2=np.std(sample2)
print(std2)
```

```python
#if change the sample size to 1000, the mean difference and variance of the
#two new samples will be smaller, if reduing the sample size to 20, the differences 
#of mean and variance will be bigger
```

```python
sample3=np.random.choice(pop1,1000,replace =True)
sample4=np.random.choice(pop2,1000,replace=True)
plt.hist(sample3,alpha=0.3,label='sample3')
plt.hist(sample4,alpha=0.3,label='sample4')
plt.legend(loc='upper right')
plt.show()
```

```python
mean3=np.mean(sample3)
print(mean3)
std3=np.std(sample3)
print(std3)

mean4=np.mean(sample4)
print(mean4)
std4=np.std(sample4)
print(std4)
```

```python
# real result is, difference between mean is smaller,but the std difference
# are still very obsvious and big, didn't change much
#and the variance difference became bigger, why??
```

```python
newpop1=np.random.binomial(10,0.3,10000)
pop2=np.random.binomial(10,0.5,10000)

plt.hist(pop1,alpha=0.3,label='newpop1')
plt.hist(pop2,alpha=0.3,label='pop2')   
plt.legend(loc='upper right')
plt.show()
```

```python
newsample1=np.random.choice(newpop1,100,replace =True)
sample2=np.random.choice(pop2,100,replace=True)
plt.hist(newsample1,alpha=0.3,label='newsample1')
plt.hist(sample2,alpha=0.3,label='sample2')
plt.legend(loc='upper right')
plt.show()
```

```python
print(newsample1.mean())
print(sample2.mean())
print(newsample1.std())
print(sample2.std())

# Compute the difference between the two sample means.
diff=sample2.mean( ) -newsample1.mean()
print(diff)
```

```python
size=np.array([len(newsample1),len(sample2)])
print(size)
sd=np.array([newsample1.std(),sample2.std()])
print(sd)
```

```python
#the square root of the sum
sqrt=(sum(sd**2/size))**0.5
#t value
tvalue=diff/sqrt
print(tvalue)
```

```python
#change group1 p to 1 

changepop1=np.random.binomial(10,0.4,10000)
pop2=np.random.binomial(10,0.5,10000)

plt.hist(changepop1,alpha=0.3,label='changepop1')
plt.hist(pop2,alpha=0.3,label='pop2')   
plt.legend(loc='upper right')
plt.show()

changesample1=np.random.choice(changepop1,100,replace =True)
sample2=np.random.choice(pop2,100,replace=True)
plt.hist(changesample1,alpha=0.3,label='changesample1')
plt.hist(sample2,alpha=0.3,label='sample2')
plt.legend(loc='upper right')
plt.show()

print(changesample1.mean())
print(sample2.mean())
print(changesample1.std())
print(sample2.std())

# Compute the difference between the two sample means.
changediff=sample2.mean( ) -changesample1.mean()
print(changediff)

changesize=np.array([len(changesample1),len(sample2)])
print(changesize)
changesd=np.array([changesample1.std(),sample2.std()])
print(changesd)

#the square root of the sum
changesqrt=(sum(changesd**2/changesize))**0.5
#t value
changetvalue=changediff/changesqrt
print(changetvalue)
```

```python
#t-value became smaller and smaller after twice change of p-value in the binomial distribution pop
```

```python
#change the two pops to normal distribution and gamma distribution
rand1=np.random.normal(10,1,10000)
rand2=np.random.gamma(5,2,10000)

plt.hist(rand1,alpha=0.3,label='rand1')
plt.hist(rand2,alpha=0.3,label='rand2')   
plt.legend(loc='upper right')
plt.show()

```

```python
randsample1=np.random.choice(rand1,100,replace =True)
randsample2=np.random.choice(rand2,100,replace=True)
plt.hist(rand1,alpha=0.3,label='rand1')
plt.hist(rand2,alpha=0.3,label='rand2')
plt.legend(loc='upper right')
plt.show()
```

```python
print(randsample1.mean())
print(randsample2.mean())
print(randsample1.std())
print(randsample2.std())

# Compute the difference between the two sample means.
randsamplediff=randsample2.mean( ) - randsample1.mean()
print(randsamplediff)

randsamplesize=np.array([len(randsample1),len(randsample2)])
print(randsamplesize)
randsd=np.array([randsample1.std(),randsample2.std()])
print(randsd)

#the square root of the sum
randsqrt=(sum(randsd**2/randsamplesize))**0.5
#t value
randtvalue=randsamplediff/randsqrt
print(randtvalue)
```

```python
#smaller the t-value, the smaller difference between two pops
```

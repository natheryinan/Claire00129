```python
#Question 3 What is the mean movie length? Calculate this by defining a function.

sql_3='''
SELECT length
FROM film
'''

lengths=engine.execute(sql_3)
engine.dispose()
rows3=lengths.fetchall()
print(rows3)
type(rows3)
lengths = [x[0] for x in rows3]
print(lengths)
```

```python
# Calculate mean of the movie length

def mean(numbers):
  total_sum = 0;  # initialize the sum to zero
  for n in numbers:
    total_sum += n # add up the numbers in the array
  count = len(numbers)   # find the length of the array
  avg = total_sum / count  # calculate the mean
  return avg  # return the result

print('The mean of the movie lengths is {}'.format(mean(lengths)))
```

```python
#Question 4  What is the median movie length? Calculate this by defining a function.
import math

def median(numbers):
    numbers.sort()
    count=len(numbers)
    is_even = count % 2 == 0 #result is true or false
    
    if is_even==True:
      middle=math.floor(count/2)
      a=numbers[middle]
      b=numbers[middle+1]
      avg=(a+b)/2
    else:
        avg=numbers[math.floor(count/2)]
    return avg
    

print('The median of the charges is {}'.format(median(lengths)))
```

```python
# Question 5 Calculate the standard deviation of the movie lengths. Calculate this by defining a function.
def std(numbers):
  mean_length=mean(numbers)
  length_number=len(numbers)
  sum_diff=0
  for i in numbers:
    diff=i-mean_length
    squared=math.pow(diff,2)
    sum_diff+=squared
  std=math.sqrt(sum_diff/(length_number-1))
  return std
    

print('The standard deviation of the charges is {}'.format(std(lengths)))
```

According to the results above, the most 5 common Journal titles are Plos One, Journal of Biological chemistry, 
neuroimage,nucleic acids research, plos pathogens

```python
# count total articles for top 5 journal 
count1= wellcome_new['Journal title'].value_counts()
print(count1)
```

```python
# count total articles for top 5 journal 
top1=wellcome_new.loc[wellcome_new['Journal title'] == 'plos one', 'Article title'].count()
top2=wellcome_new.loc[wellcome_new['Journal title'] == 'journal of biological chemistry', 'Article title'].count()
top3=wellcome_new.loc[wellcome_new['Journal title'] == 'neuroimage', 'Article title'].count()
top4=wellcome_new.loc[wellcome_new['Journal title'] == 'nucleic acids research', 'Article title'].count()
top5=wellcome_new.loc[wellcome_new['Journal title'] == 'plos genetics', 'Article title'].count()
print(top1)
print(top2)
print(top3)
print(top4)
print(top5)
```

```python
# calculate mean, median, std to column Cost
wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'] = wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'].apply(lambda x: ''.join(list(filter(str.isdigit, str(x)))))

print(wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'].head(10))# need to divide them by 100
wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'] = pd.to_numeric(wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'], errors='coerce')


print(wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'].head(20))
wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)']=wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)']/100
print(wellcome_new['COST (£) charged to Wellcome (inc VAT when charged)'].head(20))
```

```python
wellcome_new.groupby(wellcome_new['Journal title'])['COST (£) charged to Wellcome (inc VAT when charged)'].mean()
```

```python
wellcome_new.groupby(wellcome_new['Journal title'])['COST (£) charged to Wellcome (inc VAT when charged)'].median()
```

```python
wellcome_new.groupby(wellcome_new['Journal title'])['COST (£) charged to Wellcome (inc VAT when charged)'].std()
```

```python
lendingdata = pd.read_csv('data/lendingclub/accepted_2007_to_2018Q4.csv')
lendingdata.head(20)
```

```python
type(lendingdata)
```

```python
#make the dataset smaller to reduce the running time, running time on original dataset is too long
lendingdatanew=lendingdata.iloc[:10000,:25]
```

```python
lendingdatanew.info()
```

```python
annual_income=lendingdatanew['annual_inc']
print(annual_income)
```

```python
annual_income = annual_income.dropna()
```

```python
annual_income.count()
```

```python
plt.figure(figsize=(18, 12))
sns.distplot(annual_income, kde=False, rug=True);
```

```python
# The histogram above show the most of the data points in this reduced dataset are centered between 0-30000 annual_income,and
# there are a few outliers according to the rug in dark blue, and this plot show the column is not normally distributed obviously
# But comparing to boxplot, this histogram is not that clear showing outliers.
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))
ax = sns.boxplot(x=annual_income)
```

```python
# The box plot above is very good for showing outliers, all the points showing to right side of the right side line, which are all outliers
```

```python
income_below = annual_income[annual_income <= 100000]
income_above = annual_income[annual_income >= 100000]
print(income_below)
print(income_above)

```

```python
#histogram plots on income_below and income_above
print(annual_income.max())

sns.set(style="white")

fig, ax = plt.subplots()
for a in [income_above,income_below]:
    sns.distplot(a, bins=range(1, 500000, 1000), ax=ax,kde=False)
```

```python
# From the plot above I can see the two groups of data plots has separate line at 10000.The two plots are not normally distributed.
#group annual income under 10000, has more data than the group annual income more than 10000, and the most people from group annual 
#income below 10000,has income around 40000
# Because dataset has some high value outliers, so it cause this plot not that clear of the two histogram
```

```python
# violin plot on annual_income this variable
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))
ax = sns.violinplot(x=annual_income)
```

```python
# This plot did show majority data points are ranging from 0-30000, but it didn't show much information about outliers
```

```python
# two continuous variables and plots,plot three different ways
# Chose int_rate and loan_amnt 
x=lendingdatanew['int_rate']
y=lendingdatanew['installment']
plt.figure(figsize=(18, 12))
ax = sns.scatterplot(x, y, data=lendingdatanew)
```

```python
# The int_rate and installments these two variables do have some pattern showing from this scatter plot above. There are certain
#values of int_rate has many data plots, each int_rate value has a bunch of installments values, but when int_rate higher than 20%
# the number of installments values become less, when int_rate higher than 25%, the installments are very few comparing to the lower int_rate
# which makes much sense according to the finance common sense
```

```python
plt.figure(figsize=(18, 12))
plt.hist(lendingdatanew['int_rate'], color='red',  alpha=.5,label='int_rate')  # alpha just controls the opacity
plt.hist(lendingdatanew['installment'], color='blue', alpha=.5,label='installment')
plt.xlabel('installment')
plt.legend(loc='upper right')
plt.title('plot')
plt.show()

# The plot below shows installments data points are not normally disributed and int_rate data plots are either.
# This histogram is not very informative abotu comparing these two columns.
```

```python
#joint plot of two variables
g = sns.jointplot(x, y, data=lendingdatanew)
# This jointplot below has benefit showing that installments and int_rate are not normally distributed, and interest_rate between 
#6%-20% are very common, and installments values commonly ranging from 50-800
```

```python
with sns.axes_style("white"):
    sns.jointplot(x, y, kind="hex", color="k");
    # This hexplot below is very pleasentlly visualized. It has benefits of showing relationships between int_rate and installments
    # There are a few spots on plot showing two columns has strong relationships, like when int_rate=8%, and installment=350; when 
    #int_rate=12.5%,installments =300, this is a very informative plot to show realtionships.
```

```python
sns.jointplot(x, y, data=lendingdatanew, kind="kde");
# Same as hex plot above, this kernel density plot.
# int_rate raning from 7-15%, and installments ranging from 50-400, are the most common values in these two columns.Good indicator 
# for future research
```

```python
# calculate age of all patients rows in dataset encounters_patients
from datetime import date

# def calculate_age(born):
#     encounter_time = date.encounters_patients['START_x']
#     return encounter_time.year - born.year - ((encounter_time.month, encounter.day) < (born.month, born.day))
                                     
import pandas as pd
from io import StringIO



encounters_patients['Age'] = pd.to_datetime(encounters_patients['START_x']) - pd.to_datetime(encounters_patients['BIRTHDATE'])

encounters_patients['Age'] = encounters_patients['Age'] /  np.timedelta64(1, 'Y')
encounters_patients['Age'] = round(encounters_patients['Age'])


print(encounters_patients['Age'])
```

```python
# readmission 

readmission_df=encounters_patients[['encounter_Id','START_x','patient_Id']].copy()
print(readmission_df.head(10))
```

```python
count_encounter = readmission_df.groupby('patient_Id')['encounter_Id'].count()
count_encounter
```

```python
#readmission_agg = readmission_df.groupby(['patient_Id']).sort(['START_x'])

# readmission_df.groupby('patient_Id').apply(lambda x: x.sort_values('START_x'))
# readmission_df


readmission_df['readmission_diff30'] = readmission_df.sort_values(['patient_Id','START_x']).groupby('patient_Id')['START_x'].diff()
readmission_df['readmission_diff30'] 
```

```python
readmission_df['readmission_diff30'] = readmission_df['readmission_diff30'].dt.days
```

```python
readmission_df['readmission_diff30']
```

```python
readmission_df.loc[(readmission_df['readmission_diff30'] <30) & (readmission_df['readmission_diff30'] >0), 'readmission_diff30'] = 1
readmission_df.loc[(readmission_df['readmission_diff30'] >= 30) , 'readmission_diff30'] = 0
```

```python
readmission_df['readmission_diff30']
```

```python
readmission_df['readmission_diff30'] = readmission_df['readmission_diff30'].fillna(0)
```

```python
readmission_df['readmission_diff30']
```

```python
# readmission 90 days

readmission_df['readmission_diff90'] = readmission_df.sort_values(['patient_Id','START_x']).groupby('patient_Id')['START_x'].diff()
readmission_df['readmission_diff90'] 
```

```python
readmission_df['readmission_diff90'] = readmission_df['readmission_diff90'].dt.days
```

```python
readmission_df.loc[(readmission_df['readmission_diff90'] <90) & (readmission_df['readmission_diff90'] >0), 'readmission_diff90'] = 1
readmission_df.loc[(readmission_df['readmission_diff90'] >= 90) , 'readmission_diff90'] = 0
```

```python
readmission_df['readmission_diff90'] = readmission_df['readmission_diff90'].fillna(0)
```

```python
readmission_df['readmission_diff90']
```

```python
# Date of the first readmission for drug overdose within 90 days. Leave N/A if no readmissions for drug overdose 
# within 90 days.
#readmission_df.loc[readmission_df.groupby('').date.idxmax()]

grouped=readmission_df.groupby('patient_Id')['START_x']
readmission_df['firsttime_readmission']= grouped.transform(lambda x: x.nsmallest(2).max())
```

```python
readmission_df['firsttime_readmission']
```

```python
#readmission_df['first_encounter']=readmission_df.loc[readmission_df.groupby('patient_Id')['START_x'].idxmax()]

grouped=readmission_df.groupby('patient_Id')['START_x']
readmission_df['first_encounter']= grouped.transform(lambda x: x.min())
```

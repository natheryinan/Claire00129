```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
distributionvar=np.random.poisson(5,100)
```

```python
plt.hist(distributionvar)
plt.show()
```

```python
plt.hist(distributionvar)
plt.axvline(distributionvar.mean(),color='y',linestyle='solid',linewidth=2)
plt.axvline(distributionvar.mean() + distributionvar.std(),color='y',linestyle='dashed',linewidth=2)
plt.axvline(distributionvar.mean() - distributionvar.std(),color='y',linestyle='dashed',linewidth=2)
a=np.mean(distributionvar)
print(a)
b=np.var(distributionvar)
print(b)
```

```python
norm1=np.random.normal(5,0.5,100)
plt.hist(norm1,bins=20,color='y')
plt.show()
norm2=np.random.normal(10,1,100)
plt.hist(norm2,bins=20,color='y')
plt.show()
norm3=norm1+norm2
plt.hist(norm3,bins=20,color='b')
plt.show()
```

```python
plt.hist(norm3,bins=20,color='b')
plt.axvline(norm3.mean(),color='y',linestyle='solid',linewidth=2)
plt.axvline(norm3.mean() + distributionvar.std(),color='y',linestyle='dashed',linewidth=2)
plt.axvline(norm3.mean() - distributionvar.std(),color='y',linestyle='dashed',linewidth=2)
a1=np.mean(norm3)
print(a1)
b1=np.var(norm3)
print(b1)
```
